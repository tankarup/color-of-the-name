```python
import csv
import pprint

files = ['color_wa.csv', 'color_you.csv']

colors = []
index = 0

for file in files:
    color = {}
    with open(file, 'r',encoding="cp932") as f:
        rows = csv.reader(f)
        header = next(rows)
        for row in rows:
            color[row[0]] = {'名前': row[1], 'よみ': row[2]}
    colors.append(color)
```
